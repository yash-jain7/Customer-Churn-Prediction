### Visualization

```python
#Helper function
def stacked_graph(data, feature):
    phone_service = data.groupby([feature, 'Churn']).size().unstack()
    phone_service = phone_service.div(phone_service.sum(axis = 1), axis = 0) * 100
    phone_service.reset_index(inplace = True)
    phone_service.plot(x = feature, kind = 'barh', stacked = True)
    plt.xlabel('percent')
    plt.ylabel(feature)
    plt.show()
```

### Feature Engineering

```python
data.drop(['customerID'], axis = 1, inplace = True)

data['gender'] = data['gender'].replace('Female', 1)
data['gender'] = data['gender'].replace('Male', 0)

data['Partner'] = data['Partner'].replace('Yes', 1)
data['Partner'] = data['Partner'].replace('No', 0)

data['Dependents'] = data['Dependents'].replace('Yes', 1)
data['Dependents'] = data['Dependents'].replace('No', 0)

data['PhoneService'] = data['PhoneService'].replace('Yes', 1)
data['PhoneService'] = data['PhoneService'].replace('No', 0)

data['PaperlessBilling'] = data['PaperlessBilling'].replace('Yes', 1)
data['PaperlessBilling'] = data['PaperlessBilling'].replace('No', 0)

data['Churn'] = data['Churn'].replace('Yes', 1)
data['Churn'] = data['Churn'].replace('No', 0)
```

```python
data = pd.get_dummies(data, columns = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'DeviceProtection', 'TechSupport',
                                  'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod', 'OnlineBackup'])
```

```python
X = data.drop(['Churn'], axis = 1)
y = data['Churn']
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, stratify = y, shuffle = True, random_state = 0)
```

```python
sc = StandardScaler()
X_train[['tenure', 'MonthlyCharges', 'TotalCharges']] = sc.fit_transform(X_train[['tenure', 'MonthlyCharges', 'TotalCharges']])
X_test[['tenure', 'MonthlyCharges', 'TotalCharges']] = sc.transform(X_test[['tenure', 'MonthlyCharges', 'TotalCharges']])
```

# Creating Oversampled Dataset

```python
ros = RandomOverSampler(random_state=0)
X_over, y_over = ros.fit_resample(X_train, y_train)
```

# Creating Undersampled Dataset

```python
rus = RandomUnderSampler(random_state=0)
X_under, y_under = rus.fit_resample(X_train, y_train)
```

# Logistic regression

### No sampling

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
cm = confusion_matrix(y_test, y_pred, normalize = 'true')
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.show()
print("Precision score:", precision_score(y_test, y_pred))
print("Recall score:", recall_score(y_test, y_pred))
print("f1 score:", f1_score(y_test, y_pred))
```

### Oversampling

```python
lr = LogisticRegression()
lr.fit(X_over, y_over)
y_pred = lr.predict(X_test)
cm = confusion_matrix(y_test, y_pred, normalize = 'true')
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.show()
print("Precision score:", precision_score(y_test, y_pred))
print("Recall score:", recall_score(y_test, y_pred))
print("f1 score:", f1_score(y_test, y_pred))
```

### Undersampling

```python
lr = LogisticRegression()
lr.fit(X_under, y_under)
y_pred = lr.predict(X_test)
cm = confusion_matrix(y_test, y_pred, normalize = 'true')
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.show()
print("Precision score:", precision_score(y_test, y_pred))
print("Recall score:", recall_score(y_test, y_pred))
print("f1 score:", f1_score(y_test, y_pred))
```
